```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
load_df = pd.read_csv('../data/load_raw.csv')
load_df.head()
```

```python
load_df.info()
```

```python
load_df['Date'] = pd.to_datetime(load_df[['year', 'month', 'day']])
load_df.head()
```

```python
load_df.set_index('Date', inplace=True)
load_df.head()
```

```python
hour_cols = [f'h{i}' for i in range(1, 25)]

load_df[hour_cols] = load_df[hour_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
load_df['Daily_Load'] = load_df[hour_cols].sum(axis=1)
load_df[['Daily_Load']].head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(load_df.index, load_df['Daily_Load'])
plt.title("Total Daily Electricity Load Over Time")
plt.xlabel("Date")
plt.ylabel("Daily Load")
plt.show()
```

```python
hourly_mean = load_df[hour_cols].mean()

plt.figure(figsize=(10,4))
hourly_mean.plot(kind='bar')
plt.title("Average Electricity Load by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Load")
plt.show()
```

```python
load_df['DayOfWeek'] = load_df.index.dayofweek
load_df['Is_Weekend'] = load_df['DayOfWeek'] >= 5

weekday_load = load_df[~load_df['Is_Weekend']]['Daily_Load']
weekend_load = load_df[load_df['Is_Weekend']]['Daily_Load']

plt.figure(figsize=(6,4))
sns.boxplot(data=[weekday_load, weekend_load])
plt.xticks([0,1], ['Weekday', 'Weekend'])
plt.title("Weekday vs Weekend Electricity Usage")
plt.ylabel("Daily Load")
plt.show()
```

```python
load_df['Month'] = load_df.index.month

monthly_avg = load_df.groupby('Month')['Daily_Load'].mean()

plt.figure(figsize=(8,4))
monthly_avg.plot(kind='bar')
plt.title("Average Monthly Electricity Load")
plt.xlabel("Month")
plt.ylabel("Average Daily Load")
plt.show()
```
